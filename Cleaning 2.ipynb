```python
df_split_date = df_main.copy()
```

```python
# Split 'DATE OCC' into year, month and day.
df_split_date['DATE OCC'] = pd.to_datetime(df_split_date['DATE OCC'])
df_split_date['YEAR OCC'] = df_split_date['DATE OCC'].dt.year
df_split_date['MONTH OCC'] = df_split_date['DATE OCC'].dt.month
df_split_date['DAY OCC'] = df_split_date['DATE OCC'].dt.day
df_split_date
```
